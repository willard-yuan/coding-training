Leetcode202: Happy Number

## 题目描述

来源：[Happy Number](https://leetcode.com/problems/happy-number/)

>A happy number is a number defined by the following process: Starting with any positive integer, replace the number by the sum of the squares of its digits, and repeat the process until the number equals 1 (where it will stay), or it loops endlessly in a cycle which does not include 1. Those numbers for which this process ends in 1 are happy numbers.

>**Example**: 19 is a happy number

>1^2 + 9^2 = 82

>8^2 + 2^2 = 68

>6^2 + 8^2 = 100

>1^2 + 0^2 + 0^2 = 1

## 分析

> 给定任一一个正数，将它的各个位平方后相加，得到新数后重复平方相加的过程，这个过程有可能得到两种结果：一种是最后得到的新数为1；另外一种是陷入循环但新数永远不为1。将得到为1的数成为"happy number"。

> C++解题思路：首先写出求和过程`addSum()`，另外我们每次将计算的未重复的新数放入`set<int>SqList`中，当进行下一个过程得到一个新数时，我们判断`set<int>SqList`中是否包含该数，如果包含，这终止后面的计算，并将其判断为不是"happy number"。

> Python处理思路：Python处理思路和C++处理的过程一样，只不过我们用一个列表`sums`来保存计算的未重复的新数。


## 代码

```cpp
#include<iostream>
#include<set>
using namespace std;

class Solution {
public:
	int addSum(int n)
	{
		int sum = 0,d =0,flag = 0, n1= n;
		while(n1)
		{
			n1/=10;
			flag++;
		}
		for(int i=flag;i>0;i--)
		{
			d = n%10;
			n/=10;
			sum = sum+d*d;
		}
		return sum;
	}

    bool isHappy(int n)
	{
		set<int>SqList;
		int sumTmp = addSum(n);
        while(sumTmp!=1&&SqList.find(sumTmp)==SqList.end())
        {
			SqList.insert(sumTmp);
			sumTmp = addSum(sumTmp);
        }
		if (sumTmp ==1)
			return true;
		else
			return false;
    }
};

int main()
{
	int n = 19;
	Solution Solver;
	bool str = Solver.isHappy(n);
	cout<<n<<" is "<<str<<endl;
	system("pause");
}
```

````python
class Solution:
    # @param {integer} n
    # @return {boolean}
    def __init__(self):
        pass

    def addSum(self, num):
        sum1 = 0
        for i, bit in enumerate(str(num)):
            sum1 += int(bit) * int(bit)
        return sum1

    def isHappy(self, n):
        sums = []
        Tmp = self.addSum(n)
        if Tmp == 1:
            return True
        else:
            sums.append(Tmp)
        Tmp = self.addSum(Tmp)
        #while bool(Tmp != 1) & bool(Tmp not in sums):
        while (Tmp != 1) & (Tmp not in sums):
            sums.append(Tmp)
            Tmp = self.addSum(Tmp)
        if Tmp == 1:
            return True
        else:
            return False


if __name__ == '__main__':
    n = 24
    solver = Solution()
    boolTmp = solver.isHappy(n)
    print boolTmp
````
